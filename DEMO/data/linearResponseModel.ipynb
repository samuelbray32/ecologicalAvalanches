## Linear Response Model

Use to generate data from the model

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
#generate data
beta=np.array([.1,.01,.001,.0001,.00001]) #lambda values from the paper equation
N=np.random.normal(1,.01,3*1000000)
x=np.zeros((N.size,beta.size))
x[0,:]=N[0]

for i in range(1,N.size):
    x[i,:]=N[i]*(x[i-1,:]+beta*(1-x[i-1,:]))
```

```python
#plot curves
fig=plt.figure(figsize=(10,5))
for i in range(len(beta)):
    plt.plot(x[::10,i],lw=.7, label=np.log10(beta[i]),color=cm.plasma(i/5.0))
plt.xlabel('time')
plt.ylabel('x')
plt.legend()
plt.show()
```

```python
# save data for analysis
np.save('linModel.npy',x)
```
